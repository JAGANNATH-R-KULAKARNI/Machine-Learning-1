```python
y= np.array([3, 8, 1, 10, 5, 7])
plt.plot(y)
plt.show()
# default of y is 0,1,2,3,4,5
```

```python
plt.plot(y,marker='o')
plt.show()
```

```python
plt.plot(y,marker='*')
plt.show()
```

```python
plt.xlabel("Average Pulse")
plt.title("Sports Watch Data")
plt.ylabel("Calorie Burnage")
plt.plot(y,marker='*',color='r')
plt.show()
```

```python
plt.xlabel("Average Pulse")
plt.title("Sports Watch Data")
plt.ylabel("Calorie Burnage")
plt.grid()
plt.plot(y,marker='*',color='b')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```
